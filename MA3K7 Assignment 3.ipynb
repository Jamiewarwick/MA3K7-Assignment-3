```python
import numpy as np
import matplotlib.pyplot as plt 
import scipy.stats as ss
from collections import deque
```

```python
#random game of numbers in the hat, defined recursively
```

```python
def recursive_hat(X):    
    if len(X)==1:
        return X #return last value
    else:
        np.random.shuffle(X) #randomise the numbers in the hat
        #two_numbs = [X[0], X[1]]
        diff = np.abs(X[0]-X[1]) #calculate difference of first two and remove them
        X.popleft()
        X.popleft()
        X.append(diff) #append difference
        return recursive_hat(X) #repeat
```

Simulating one game for numbers 1 to 20

```python
X= np.arange(1,21) #numbers 1 to 20
d = deque(X)
recursive_hat(d)[0]
```

```python
#Lets simulate 500 games with 100 numbers in the hats:
```

```python
final_vals = []
trials = 500
hat_max = 100
for i in range(trials):
    X= np.arange(1,hat_max+1)
    d = deque(X)
    #y = recursive_hat(d)[0]
    final_vals.append(recursive_hat(d)[0])
    #print(y)
    
print(final_vals)
    
```

And now let's plot the distributions.

```python
plt.hist(final_vals, bins=20) #for 2025
```

```python
#plot for 100 and 101
final_vals100 = []
trials = 50000
hat_max = 100
for i in range(trials):
    X= np.arange(1,hat_max+1)
    d = deque(X)
    #y = recursive_hat(d)[0]
    final_vals100.append(recursive_hat(d)[0])
    #print(y)
final_vals101 = []
trials = 50000
hat_max = 101
for i in range(trials):
    X= np.arange(1,hat_max+1)
    d = deque(X)
    #y = recursive_hat(d)[0]
    final_vals101.append(recursive_hat(d)[0])
    #print(y)
    
    
plt.hist(final_vals100, bins=25,density = True, alpha = 0.7, label = "100 in hat")
plt.hist(final_vals101, bins=25,density = True, alpha = 0.7, label = "101 in hat")
plt.title("Probability final number is $n$, starting with 100 and 101 numbers")
plt.legend()
plt.xlabel("$n$")
plt.ylabel("Probability")
plt.show()
```

And we see this nonlinear decay, as $n$ approaches $N$ ie 100  and 101 here, the probability of $n$ occuring approaches 0

this decay appears to be quadratic or exponential

```python
#plot for 2025
final_vals2025 = []
trials = 800
hat_max = 2025
for i in range(trials):
    X= np.arange(1,hat_max+1)
    d = deque(X)
    #y = recursive_hat(d)[0]
    final_vals2025.append(recursive_hat(d)[0])
    #print(y)
    
    
plt.hist(final_vals2025, bins=20,density = True)
plt.title("Probability final number is $n$")
plt.xlabel("$n$")
plt.ylabel("Probability")
plt.show()
```

```python
#lets fit an exponential distribution
```

```python
data2025 = np.array(final_vals2025)
x = np.linspace(1,2000,100000)

P = ss.expon.fit(data2025)
print(P)
#fit exponential distribution

rP = ss.expon.pdf(x, *P)
plt.hist(data2025, density=True, bins=25, label = "2025 in hat")
plt.title("Probability final number is $n$, starting with 2025 numbers")
plt.plot(x, rP, label = "exponential fit")
plt.xlabel("$n$")
plt.ylabel("Probability")
plt.legend()
plt.show()
```

And as we see this is quite a good fit for the data
